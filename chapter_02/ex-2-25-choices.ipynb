# Choice rules


```python
%%file ./instances/choice_with_local.lp
p(a;b).
q(1..4).
1 {r(X,Y): p(X)} :- q(Y).
#show r/2.
```

There are 81 models: for each q=1..4 we can choose at least one p(a) or p(b). More precisely for each q we can choose from

$\{r(a,q)\}$, $\{r(b,q)\}$ or $\{r(a,q), r(b,q) \}$

Thus three sets for each q. In total $4^3$=81

Rewriting the above program without the use of local variables, i.e. p(X):

```python
%%file ./instances/choice_without_local.lp
p(a;b).
q(1..4).
1 {r(a,Y; b,Y)} :- q(Y).

#show r/2.
```
